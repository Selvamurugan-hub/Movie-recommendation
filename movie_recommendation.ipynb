## Project Description

This project is a simple Tamil Movie Recommendation System developed using Python in Google Colab.
It recommends popular Tamil movies to users based on their preferred genre.

The project demonstrates basic Python programming concepts such as:
- Lists and dictionaries
- Functions
- Conditional statements
- User input handling


```python
# Tamil Movie Recommendation System with Ajith Kumar Movies

tamil_movies = {
    "Action": [
        "Vikram",
        "Thuppakki",
        "Master",
        "Vedalam",
        "Viswasam",
        "Valimai"
    ],
    "Comedy": [
        "Boss Engira Bhaskaran",
        "Panchathanthiram",
        "Soodhu Kavvum"
    ],
    "Drama": [
        "Asuran",
        "Jai Bhim",
        "Pariyerum Perumal"
    ],
    "Romance": [
        "96",
        "Vinnaithaandi Varuvaayaa",
        "Alaipayuthey",
        "Kadhal Kottai"
    ],
    "Thriller": [
        "Ratsasan",
        "Dhuruvangal Pathinaaru",
        "Mankatha",
        "Yennai Arindhaal"
    ]
}

def recommend_movie():
    print("Available Genres:\n")
    for genre in tamil_movies:
        print("-", genre)

    choice = input("\nEnter your preferred genre: ").title()

    if choice in tamil_movies:
        print("\n Recommended Tamil Movies:")
        for movie in tamil_movies[choice]:
            print("â€¢", movie)
    else:
        print("\n Sorry, genre not found.")

recommend_movie()
```
